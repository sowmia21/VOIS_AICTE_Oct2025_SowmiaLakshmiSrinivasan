```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')
sns.set_palette("Set2")
```

```python
file_path = r"C:\Users\srini\Downloads\1730285881-Airbnb_Open_Data.xlsx"
df = pd.read_excel(file_path)

```

```python
df.isnull().sum()
```

```python
df.duplicated().value_counts() 
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop(['house_rules', 'license'], axis=1, inplace=True)
```

```python
categorical_cols = ['NAME', 'host_identity_verified', 'host name', 
                    'neighbourhood group', 'neighbourhood', 
                    'country', 'country code', 'instant_bookable', 'cancellation_policy']

for col in categorical_cols:
    df[col] = df[col].fillna('Unknown')

```

```python
df['price'] = df['price'].fillna(df['price'].median())
df['minimum nights'] = df['minimum nights'].fillna(df['minimum nights'].median())
df['reviews per month'] = df['reviews per month'].fillna(0)
```

```python
df.isnull().sum()
```

```python
# Fill latitude & longitude
df['lat'] = df['lat'].fillna(df['lat'].median())
df['long'] = df['long'].fillna(df['long'].median())

# Fill Construction year
df['Construction year'] = df['Construction year'].fillna(0)

# Fill service fee
df['service fee'] = df['service fee'].fillna(df['service fee'].median())

# Fill number of reviews
df['number of reviews'] = df['number of reviews'].fillna(0)

# Fill last review
df['last review'] = df['last review'].fillna('No Review')
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')

# Fill review rate number
df['review rate number'] = df['review rate number'].fillna(df['review rate number'].median())

# Fill calculated host listings count
df['calculated host listings count'] = df['calculated host listings count'].fillna(0)

# Fill availability 365
df['availability 365'] = df['availability 365'].fillna(0)
```

```python
df.isnull().sum()
```

```python
# Convert to datetime but keep missing as NaT
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
```

```python
df.duplicated().sum()
```

```python
# Number of listings
print("Total listings:", df.shape[0])

# Number of hosts
print("Total hosts:", df['host id'].nunique())

# Listings by neighbourhood group
print(df['neighbourhood group'].value_counts())
```

```python
df['neighbourhood group'] = df['neighbourhood group'].replace({
    'brookln': 'Brooklyn',
    'manhatan': 'Manhattan',
    'Unknown': np.nan  # Optional: set Unknown to NaN if needed
})
```

```python
df['neighbourhood group'].value_counts()
```

```python
df['price'] = df['price'].astype(str).str.replace('$','').str.replace(',','').astype(float)
df['service fee'] = df['service fee'].astype(str).str.replace('$','').str.replace(',','').astype(float)
```

```python
df.rename(columns={'price': 'price($)', 'service fee': 'service_fee($)'}, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
Q1 = df['availability 365'].quantile(0.25)
Q3 = df['availability 365'].quantile(0.75)
IQR = Q3 - Q1

df = df[~((df['availability 365'] < (Q1 - 1.5*IQR)) | (df['availability 365'] > (Q3 + 1.5*IQR)))]
```

```python
# Check for duplicates 
dup_count = df.duplicated().sum()
print("Duplicate rows:", dup_count)

# Check for missing values 
print("\nMissing values per column:\n")
print(df.isnull().sum())

# Check data types 
print("\nData types:\n")
print(df.info())

# Check unique values in key categorical columns 
categorical_cols = ['neighbourhood group', 'room type', 'country', 'instant_bookable', 'cancellation_policy']
for col in categorical_cols:
    print(f"\nUnique values in '{col}':\n{df[col].unique()}")

# Check price and service fee formatting 
print("\nPrice and Service Fee sample:\n")
print(df[['price($)', 'service_fee($)']].head())

# Check for outliers in availability 365 
plt.figure(figsize=(8,4))
sns.boxplot(df['availability 365'])
plt.title('Outliers check: availability 365')
plt.show()
```

```python
# Count of each room type
room_counts = df['room type'].value_counts()
print("Different property types and their counts:\n", room_counts)

# Optional visualization
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='room type', order=room_counts.index)
plt.title('Distribution of Property Types')
plt.ylabel('Number of Listings')
plt.show()
```

```python
neigh_counts = df['neighbourhood group'].value_counts()
print("Number of listings per neighbourhood group:\n", neigh_counts)

# Bar plot

plt.figure(figsize=(8,5))
sns.barplot(x=neigh_counts.index, y=neigh_counts.values, color="skyblue")
plt.title('Listings per Neighborhood Group')
plt.ylabel('Number of Listings')
plt.show()

```

```python
avg_price = df.groupby('neighbourhood group')['price($)'].mean().sort_values(ascending=False)
print("Average price per neighbourhood group:\n", avg_price)

# Bar plot
avg_price.plot(kind='bar', figsize=(8,5), color='skyblue')
plt.title('Average Price by Neighborhood Group')
plt.ylabel('Average Price ($)')
plt.show()
```

```python
# Group by Construction Year and calculate average price
avg_price_by_year = df.groupby('Construction year')['price($)'].mean().to_frame()

# Plot the trend (zooming x-axis)
avg_price_by_year.plot()
plt.xlabel("Construction Year")
plt.ylabel("Average Price ($)")
plt.title("Average Price ($) for Properties in each Construction Year")
plt.xlim(1990, 2022)  # Zoom into years 1990â€“2020
plt.show()


```

```python
correlation = df[['Construction year', 'price($)']].corr()
print("Correlation between construction year and price:\n", correlation)
```

```python

# Top 10 hosts by total calculated host listings count
hosts = df.groupby("host name")['calculated host listings count'].sum().sort_values(ascending=False).nlargest(10).to_frame()

# Bar plot
plt.figure(figsize=(12,6))
bars = plt.bar(hosts.index, hosts['calculated host listings count'], edgecolor='black', linewidth=1.5)
plt.bar_label(bars, labels=hosts['calculated host listings count'], padding=3)

plt.xlabel("Host Name")
plt.ylabel("Calculated Host Listings Count")
plt.xticks(rotation=80)
plt.ylim([0, hosts['calculated host listings count'].max() + 10])  # Adjust ylim slightly above max
plt.title("Top 10 Hosts by Calculated Host Listings Count")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- Bar plot: Average review rate by host verification ---
review = df.groupby('host_identity_verified')['review rate number'].mean().to_frame()

plt.figure(figsize=(8,5))
bars = plt.bar(review.index, review['review rate number'], color=sns.color_palette()[0], edgecolor='black')
plt.bar_label(bars, labels=[f"{x:.2f}" for x in review['review rate number']], padding=4)

plt.ylim([0, 5])  # Adjust based on review scale
plt.xlabel("Host Verification Status")
plt.ylabel("Average Review Rate")
plt.title("Average Review Rate for each Verification Status")
plt.show()

# --- Box plot: Distribution of review rates by host verification ---
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='host_identity_verified', y='review rate number', color='skyblue' )

plt.xlabel("Host Verification Status")
plt.ylabel("Review Rate")
plt.title("Distribution of Review Rates by Host Verification Status")
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Regression plot: Price vs Service Fee
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='price($)', y='service_fee($)', scatter_kws={'alpha':0.5}, line_kws={'color':'red'})

plt.xlabel("Price ($)")
plt.ylabel("Service Fee ($)")
plt.title("Regression Plot: Price vs Service Fee")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Optional: correlation coefficient
corr = df['price($)'].corr(df['service_fee($)'])
print("Correlation between Price and Service Fee:", round(corr, 2))


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,10))

# Bar plot: Average review rate per neighborhood group and room type
sns.barplot(data=df, 
            x='neighbourhood group', 
            y='review rate number', 
            hue='room type', 
            errorbar=None, 
            palette='Set2')

plt.xlabel("Neighbourhood Group", fontsize=12)
plt.ylabel("Average Review Rate", fontsize=12)
plt.title("Average Review Rate for each Room/Property Type in each Neighbourhood Group", fontsize=14)
plt.legend(title="Room Type")
plt.show()
```
